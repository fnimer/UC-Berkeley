# Self-Study Colab Activity 1.1: Creating Hist, Scatter, Line, and Bar Plots

### Read US Real Estate Census Data - version 2024-11-06


```python
# import data from pandas
import pandas as pd
import os

#Make directory called plots
os.makedirs('real_estate', exist_ok=True)
#current_directory = os.getcwd()
#print(current_directory)

#create the DataFrame utilizing Pandas
census = pd.read_csv('/content/real_estate/RealEstateUnitedStates.csv')

#generate all info for this dataset
census.info()

#Remove the last line of NaN
#census = census.iloc[:-1 , :]

#Ignore any warnings
import warnings
warnings.filterwarnings('ignore')
```

###  Part One: In this exercise, generate a basic histogram.


```python
#Analize data profile

census.describe()
```

```python
#analyze content

census.head(20)
```

```python
#Generate a basic histogram and save the histogram as a png file
# Hint: use pandas histogram function.

#Average Sales Price
ax = census.plot.hist(column='Average Sales Price', bins=20, grid=True, figsize=(8,4), color='blue', alpha=0.6, rwidth=0.8, title='Average Sales Price - US Census Data', xlabel='Avg Sales Price', ylabel='Frequency', cumulative=False, density=False, orientation='vertical', logx=False, logy=False)
ax.figure.savefig('average sales price.pdf')
```

```python
#Generate a basic histogram and save the histogram as a png file
# Hint: use pandas histogram function.

#Mean Income - Current Dollars
ax = census.plot.hist(column='Mean Income - Current Dollars', bins=20, grid=True, figsize=(8,4), color='blue', alpha=0.6, rwidth=0.8, title='Mean Income - US Census Data', xlabel='Mean Income', ylabel='Frequency', cumulative=False, density=False, orientation='vertical', logx=False, logy=False)
ax.figure.savefig('mean income.pdf')
```

###  Part Two: In this exercise, generate a basic scatter plot.


```python
#Please generate a basic scatter plot utilizing Average Sales Price as the x axis and Mean Income - Current Dollars as the y axis
# Hint: use pandas scatter function to generate the chart.

scat = census.plot.scatter(x='Average Sales Price', y='Mean Income - Current Dollars', color='blue', title='Sales Price vs Mean Income - US Census Data', xlabel='Avg Sales Price', ylabel='Mean Income', figsize=(10,5), grid=True)
scat.figure.savefig('sales price vs mean income.pdf')

# ANSWER CHECK
print(scat)
```

###  Part Three: In this exercise, generate a basic lineplot.

```python
#Generate a basic line plot by Region as the x axis and Average Sales Price as the y axis using the census dataframe.
#for the Region = "South" using the Census data set.
#Note a new dataframe called censusdata has been created for you that will be used during the rest of this

#decide if I will create a subset of the dataframe later!!
#regioncensus = census.filter(items=['Region', 'Average Sales Price', 'Mean Income - Current Dollars'])
#regioncensus = regioncensus.loc[census.Region == 'Southeast', 'Average Sales Price':'Mean Income - Current Dollars']
#regioncensus.set_index('Region')

lines = census.plot.line(x='Region', y='Average Sales Price', figsize=(10,5), color='blue', title='Average Sales Price by Region', xlabel='Region', ylabel='Average Sales Price', grid=True, legend=True)
lines.figure.savefig('Region - Avg Sales Price.pdf')

# ANSWER CHECK
print(lines)
```

```python
#Generate a basic line plot by Region as the x axis and Mean Income - Current Dollars as the y axis

lines = census.plot.line(x='Average Sales Price', y='Mean Income - Current Dollars', figsize=(10,10), color='blue', title='Average Sales Price X Mean Income USD', xlabel='Average Sales Price', ylabel='Mean Income USD', grid=True, legend=True)
lines.figure.savefig('Average Sales Price X Mean Income.pdf')

# ANSWER CHECK
print(lines)
```

```python
#Generate a basic line plot by Region as the x axis and Mean Income - Current Dollars as the y axis

lines = census.plot.line(x='Region', y='Mean Income - Current Dollars', figsize=(10,5), color='blue', title='Mean Income by Region', xlabel='Region', ylabel='Mean Income USD', grid=True, legend=True)
lines.figure.savefig('Region - Mean Income.pdf')

# ANSWER CHECK
print(lines)
```

###  Part Four: In this exercise, generate a basic bar chart.

```python
#Generate a basic bar plot by Region as the x axis and Mean Income - Current Dollars as the y axis of the census dataframe.

bar = census.plot.bar(x='Region', y='Average Sales Price', color='blue', figsize=(50,50), title='Average Sales Price by Region', xlabel='Region', ylabel='Average Sales Price', grid=False)
bar.figure.savefig('Barplot Region x Average Sales Price.pdf')

# ANSWER CHECK
print(bar)
```

```python
#Generate a basic bar plot by Region as the x axis and Mean Income - Current Dollars as the y axis of the census dataframe.

bar = census.plot.bar(x='Region', y='Mean Income - Current Dollars', color='blue', figsize=(50,50), title='Mean Income by Region', xlabel='Region', ylabel='Mean Income', grid=False)
bar.figure.savefig('Barplot Region x Mean Income.pdf')

# ANSWER CHECK
print(bar)
```
